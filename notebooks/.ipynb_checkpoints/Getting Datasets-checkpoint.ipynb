## Data from Web Service

```python
def get_datasets(request):
    tuples = []
    for line in json.loads(request.text)['value']:
        url = line['url']
        name = line['name']
        request = requests.get(url)
        table = json_normalize(json.loads(request.text)['value'])
        tuples.append((name,table))
        
    tables = {key: value for (key, value) in tuples}
    print(tables.keys())
    return tables
```

### Deaths; murder and manslaughter, crime scene in The Netherlands
This table contains the number of persons died as a result of murder or manslaughter, where the crime scene is located in the Netherlands. The victims can be residents or non-residents. The data can be split by location of the crime, method, age and sex. The date of death is the criterion, the date of the act can be in the previous year. The ICD10 codes that belong to murder and manslaughter are X85-Y09.  
https://data.overheid.nl/data/dataset/deaths-murder-and-manslaughter-crime-scene-in-the-netherlands

```python
r = requests.get("http://opendata.cbs.nl/ODataApi/OData/81453ENG")
service = json.loads(r.text)
for i in service['value']:
    print (i)
```

```python
request = requests.get("http://opendata.cbs.nl/ODataApi/OData/81453ENG")
tables = get_datasets(request)
```

### Feelings of insecurity by background characteristics
Feelings of insecurity due to crime Scared to be at home alone, or open the front door, scary places etc. 1997 - 2004 Changed on December 14 2005. Frequency: Discontinued.  
https://data.overheid.nl/data/dataset/feelings-of-insecurity-by-background-characteristics

```python
r = requests.get("http://opendata.cbs.nl/ODataApi/OData/37775eng")
service = json.loads(r.text)
for i in service['value']:
    print (i)
```

```python
request = requests.get("http://opendata.cbs.nl/ODataApi/OData/37775eng")
tables = get_datasets(request)
```

```python
df_insecurity = tables['TypedDataSet']
df_insecurity = df_insecurity.loc[:, df_insecurity.columns != 'ID']
```

```python
#tables['AllPersonalCharacteristics']
```

```python
df_insecurity['KeyCharacteristics'] = [int(i) for i in df_insecurity['AllPersonalCharacteristics']]
```

```python
df_characteristics = tables['AllPersonalCharacteristics']
df_characteristics['KeyCharacteristics'] = [int(i) for i in df_characteristics['Key']]
df_characteristics = df_characteristics[['Title','CategoryGroupID','KeyCharacteristics']]
```

```python
df_insecurity = pd.merge(df_insecurity, df_characteristics, on='KeyCharacteristics', how='left')
```

```python
df_insecurity.index = df_insecurity['Periods']
```

## Create Dataset

```python
df_insecurity.to_csv('data/feelings_insecurity_characteristics.csv', sep=',', encoding= 'utf-8')
print('feelings_insecurity_characteristics.csv dataset created ')
```

### Crime; type of crime, municipality
This table contains figures on the number of registered crimes. These are broken down according to the type of crime, including figures on High Impact Crimes (theft / burglary, violent crimes, robberies, street robbery). The data is available per municipality. The crime types shown are a selection of all crime types, and do not add up to the Total number of crimes.
Data available from: 2014.  
https://data.overheid.nl/data/dataset/misdrijven-soort-misdrijf-gemeente
