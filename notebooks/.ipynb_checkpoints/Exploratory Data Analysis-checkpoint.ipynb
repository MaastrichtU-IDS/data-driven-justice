```python
df_crime = pd.read_csv('../data/crime/47004NED_UntypedDataSet_20022018_134938.csv', sep=';')
df_crime = df_crime.loc[:, df_crime.columns != 'ID']
```

```python
df_crime = pd.merge(df_crime, df_municipalities, left_on='Gemeenten', right_on='Key', how='left')
df_crime = pd.merge(df_crime, df_crimentype, left_on='SoortMisdrijf', right_on='Key', how='left')
```

```python
temp = df_crime[df_crime['Perioden'] == '2017HJ01']
temp = temp.copy()
temp.loc[:,'Misdrijven_1'] = df_crime[df_crime['Perioden'] == '2017HJ01']['Misdrijven_1'].multiply(2)
temp.loc[:,'Perioden'] = '2017JJ00'
```

```python
df_crime = df_crime.append(temp)
```

```python
df_crime = df_crime[(df_crime['Perioden'] != '2015HJ01') & (df_crime['Perioden'] != '2016HJ01') & (df_crime['Perioden'] != '2017HJ01')]
```

```python
df_crime = df_crime[['Perioden','Gemeenten','Misdrijven_1','Title_y']]
```

```python
df_crime.head()
```

```python
df_crime.columns = ['Perioden','CBScode','Crimes','CrimeType']
```

```python
df_crime = df_crime[(df_crime['CBScode'] != 'NL0000') & (df_crime['CBScode'] != 'GM0000')]
```

```python
df_crime = df_crime[(df_crime['CrimeType'] != 'Alle misdrijven') & (df_crime['CrimeType'] != 'HIC: High Impact Criminaliteit - totaal') & (df_crime['CrimeType'] != 'HIC: Diefstal/inbraak woning, totaal')]
```

```python
crimeTypes = ['  HIC: Diefstal/inbraak woning, voltooid',
       '  HIC: Diefstal/inbraak woning, pogingen',
       'HIC: Geweldsmisdrijven', 'HIC: Straatroof', 'HIC: Overvallen',
       'Aantasting openbare orde', 'Bedreiging', 'Brand/ontploffing',
       'Diefstal van brom-, snor-, fietsen',
       'Diefstal uit/vanaf motorvoertuigen',
       'Diefstal van motorvoertuigen',
       'Diefstal/inbraak bedrijven/instellingen',
       'Diefstal/inbraak garage/schuur/tuinhuis', 'Drugshandel',
       'Huisvredebreuk', 'Mishandeling', 'Openlijk geweld (persoon)',
       'Vernieling cq. zaakbeschadiging', 'Wapenhandel', 'Zakkenrollerij',
       'Zedenmisdrijf']
```

```python
##take substraction of arrays
```

```python
df_crime['Perioden'].unique()
```

```python
df_crime.head()
```

```python
df_crimes_table = pd.DataFrame()
for column in crimeTypes:
    crime = df_crime[df_crime['CrimeType'] == str(column)].groupby(['CBScode','Perioden']).sum()['Crimes']
    df_crimes_table[str(column)] = crime
```

```python
df_crimes_table.head()
```

```python
df_crimes_table.reset_index(level=['CBScode','Perioden'], inplace=True)
```

```python
df_population = pd.read_csv('../data/crime/population_netherlands.tsv', sep='\t')
```

```python
import seaborn as sns
```

```python
sns.kdeplot(df_population['Population'])
```

```python
#pop_stats = 
df_population['Population'].describe()
```

```python
pop_size = []
for mun in df_population['Population']:
    if mun < df_population['Population'].median():
        pop_size.append('Small')
    elif df_population['Population'].median():
        pop_size.append('Big')
```

```python
df_population['Municipality Size'] = pop_size
```

```python
df_crime_pop = pd.merge(df_crimes_table, df_population, on='CBScode', how='left')
```

```python
df_crime_pop.sort_values('CBScode').head()
#  HIC: Diefstal/inbraak woning, pogingen to train
```

```python
df_crime_pop.to_csv('../data/crime/crimes_municipality_ingest.tsv', sep='\t', encoding= 'utf-8')
```
